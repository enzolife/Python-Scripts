```python
from pandas import Series, DataFrame
import pandas as pd
import glob
import os
from datetime import datetime
```

```python
path ='D:\\Program Files (x86)\\ÁôæÂ∫¶‰∫ëÂêåÊ≠•Áõò\\Dropbox\\Shopee 2016.4.12\\2016.4.23 Data Visualization\\Pricing\\SG' # use your path

allFiles = glob.glob(path + "\\*.csv")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    pwd = os.getcwd()
    os.chdir(os.path.dirname(file_))
    
    df = pd.read_csv(os.path.basename(file_),index_col=None, header=0)
    list_.append(df)
frame = pd.concat(list_)
```

```python
frame['Purchased Time']= pd.to_datetime(frame['Purchased Time'])
frame['year/month'] = frame['Purchased Time'].dt.strftime('%Y-%m')
```

```python
order_by_category_by_country_pivot = frame.pivot_table('Order ID',aggfunc='count',index=['Category'],columns='Country')
order_by_category_by_country_pivot
```

```python
frame['Adjusted Purchase Time'] = pd.to_datetime(frame['Specific purchase time'], format='%d/%m/%Y, %H:%M:%S %p')
frame['Adjusted Purchase Hour'] = frame['Adjusted Purchase Time'].dt.hour
frame[['Specific purchase time', 'Adjusted Purchase Time', 'Adjusted Purchase Hour']]
```

```python
group = frame['Order ID'].groupby([frame['Country'], frame['Category'], frame['Adjusted Purchase Hour']])
output_frame = group.agg({'Order ID': 'count'}).rename(columns={'Order ID': 'Gross_Order'})
output_frame = output_frame.unstack('Adjusted Purchase Hour').reset_index()
output_frame.columns = output_frame.columns.droplevel(0)
output_frame.columns.values[0] = 'Category'
output_frame.columns.values[1] = 'Sub Category'
output_frame
```
