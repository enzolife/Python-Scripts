```python
df['Purchased on']=pd.to_datetime(df['Purchased on'])
df.index=df['Purchased on']
grouped = df['﻿Order ID'].groupby(pd.TimeGrouper(freq='M'))
order_by_month_year = grouped.count()
order_by_month_year
```

```python
order_by_month_year.plot(kind='bar',color='yellow')

plt.title("# of Cross Border Gross Order")
plt.xlabel("Year/Month")
plt.ylabel("# of Gross Order")

plt.show()
```
