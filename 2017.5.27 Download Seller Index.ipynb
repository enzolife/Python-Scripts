```python
docs = [
    {"doc":"Data Migration Information Fullfillment"}
    ]
path = "C:\\Users\\kuang\\Documents\\Python Scripts\\"
```

```python
#process doc list
for doc in docs:
    spreadsheet = gc.open(doc["doc"])
    for i, worksheet in enumerate(spreadsheet.worksheets()):
        filename = path + doc["doc"] + '-worksheet' + str(i) + '.csv'
        # newline = ''这部分，请参考evernote: 
        # For use with the csv module, the Python 3 open should also have newline='' as a parameter [ref]
        with open(filename, 'w', newline='') as f:
            writer = csv.writer(f)
            content = worksheet.get_all_values()
            for row in content:
                new_row=[]
                for record in row:
                    new_row.append(record)
                try:
                    writer.writerow(new_row)
                except (UnicodeEncodeError, UnicodeDecodeError):
                    print ("Caught unicode error")
    print ('== Finished Writing: ' + filename + ' ==')
print ('====== FINISHED ======')
```
