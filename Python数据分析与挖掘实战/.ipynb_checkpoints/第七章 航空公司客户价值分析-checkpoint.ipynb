```python
# 数据探索分析代码
import pandas as pd
```

```python
datafile = './data/air_data.csv'
resultfile = './temp/explore.xls'
```

```python
data = pd.read_csv(datafile, encoding='utf-8')
data.head()
```

```python
explore = data.describe(percentiles=[], include='all').T # 包括对数据的基本描述，percentiles参数是制定计算多少的分位数表（如1/4分位数，中位数等）
# T是转置，转置后更方便查阅
explore
```

```python
explore['null'] = len(data) - explore['count'] # describe()函数自动计算非空值数，需要手动计算空值数
explore
```

```python
explore = explore[['null', 'max', 'min']]
explore
```

```python
explore.columns = [u'空值数', u'最大值', u'最小值'] # 表头重命名
'''这里只选取部分探索结果'''
explore
```

```python
explore.to_excel(resultfile) # 导出结果
```

```python
datafile = './data/zscoredata.xls' # 需要进行标准化的数据文件
zscoredfile = './temp/zscoreddata.xls' # 标准差化后的数据储存路径文件
```

```python
# 标准化处理
data = pd.read_excel(datafile)
data.head()
```

```python
data = (data - data.mean(axis = 0)) / (data.std(axis = 0)) # 简洁的语句实现了标准化变换，类似地可以实现任何想要的变换
data.head()
```

```python
data.columns = ['Z' + i for i in data.columns] # 表头重命名
data.head()
```

```python
data.to_excel(zscoredfile, index=False)
```

```python
# K-means聚类算法
from sklearn.cluster import KMeans # 导入K均值聚类算法

inputfile = './temp/zscoreddata.xls' # 待聚类的数据文件
k = 5 # 需要进行的聚类类别数
```

```python
# 读取数据并进行聚类分析
data = pd.read_excel(inputfile) # 读取数据
data.head()
```

```python
# 调用k-means算法，进行聚类分析
kmodel = KMeans(n_clusters=k, n_jobs=4) # n_jobs是并行数，一般等于CPU数较好
kmodel.fit(data) # 训练模型
```

```python
kmodel.cluster_centers_ # 查看聚类中心
```

```python
kmodel.labels_ # 查看各样本对应的类别
```
