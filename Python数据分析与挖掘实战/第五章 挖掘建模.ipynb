```python
# 逻辑回归代码
import pandas as pd
```

```python
# 参数初始化
filename = './data/bankloan.xls'
data = pd.read_excel(filename)
x = data.iloc[:,:8].as_matrix()
y = data.iloc[:,8].as_matrix()
```

```python
data
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.linear_model import RandomizedLogisticRegression as RLR
```

```python
rlr = RLR() # 建立随即逻辑回归模型；筛选变量
rlr.fit(x, y) # 训练模型
rlr.get_support()
```

```python
print(u'通过随机逻辑回归模型筛选特征结束')
print(u'有效特征为：%s'%','.join(data.columns[rlr.get_support()]))
```

```python
x = data[data.columns[rlr.get_support()]].as_matrix() # 筛选好特征
```

```python
x
```

```python
lr = LR() # 建立逻辑回归模型
lr.fit(x, y) # 用筛选后的特恒数据来训练模型
print(u'逻辑回归模型训练结束。')
print(u'模型的平均正确率为：%s' % lr.score(x, y)) # 给出模型的平均正确率
```

```python
# 决策树算法预测销量高低代码
filename = './data/sales_data.xls'
data = pd.read_excel(filename, index_col=u'序号')
data
```

```python
# 数据是类别标签，要将它转换为数据
# 用1来表示“好”“是”“高”这三个属性；用-1来表示“坏”“否”“低”
data[data == u'好'] = 1
data[data == u'是'] = 1
data[data == u'高'] = 1
data[data != 1] = -1
data
```

```python
x = data.iloc[:,:3].as_matrix().astype(int)
x
```

```python
y = data.iloc[:,3].as_matrix().astype(int)
y
```

```python
from sklearn.tree import DecisionTreeClassifier as DTC
dtc = DTC(criterion='entropy') # 建立决策树模型，基于信息熵
dtc.fit(x, y) # 训练模型
```

```python
# 导入相关函数，可视化决策树
# 导出的结果是一个dot文件，需要安装Graphviz才能将它转换为pdf或png等格式
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
x = pd.DataFrame(x)
with open('tree.dot', 'w') as f:
    f = export_graphviz(dtc, feature_names=x.columns, out_file=f)
```

```python
# k-means聚类算法代码
inputfile = './data/consumption_data.xls'
outputfile = './temp/data_type.xls'
k = 3 # 聚类的类别
iteration = 500 # 聚类最大循环次数
data = pd.read_excel(inputfile, index_col='Id')
data_zs = 1.0 * (data - data.mean()) / data.std() # 数据标准化
data_zs
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=k, n_jobs=4, max_iter=iteration) # 分为k类，并发数4
model.fit(data_zs)
```

```python
# 简单打印结果
r1 = pd.Series(model.labels_).value_counts() # 统计各个类别的数目
r1
```

```python
r2 = pd.DataFrame(model.cluster_centers_) # 找出聚类中心
r2
```

```python
r = pd.concat([r2, r1], axis=1) # 横向连接（0是纵向），得到聚类中心对应的类别下的数目
r
```

```python
r.columns = list(data.columns) + [u'类目数目'] # 重命名表头
r
```

```python
# 详细输出原始数据及其分类
r = pd.concat([data, pd.Series(model.labels_, index=data.index)], axis=1) # 详细输出每个样本对应的类别
r
```

```python
pd.Series(model.labels_, index=data.index)
```

```python
r.columns = list(data.columns) + [u'聚类类别'] # 重命名表头
r
```

```python
r.to_excel(outputfile) # 保存结果
```

```python
# 绘制聚类后的概率密度图
def density_plot(data, title):
    import matplotlib.pyplot as plt
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    plt.figure()
    
    for i in range(len(data.iloc[0])):
        (data.iloc[:,i]).plot(kind='kde', label=data.columns[i], linewidth=2)
    plt.ylabel(u'密度')
    plt.xlabel(u'人数')
    plt.title(u'聚类类别%s各属性的密度曲线' % title)
    plt.legend()
    return plt
```

```python
def density_plot(data):
    import matplotlib.pyplot as plt
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    
    p = data.plot(kind='kde', linewidth=2, subplots=True, sharex=False)
    [p[i].set_ylabel(u'密度') for i in range(k)]
    plt.legend()
    return plt
```

```python
pic_output = './temp/pd_' # 概率密度图文件名前缀
for i in range(k):
    density_plot(data[r[u'聚类类别'] == i]).savefig(u'%s%s.png' % (pic_output, i))
```
