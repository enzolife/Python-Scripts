```python
    # 16.4.1 连接到IMAP服务器
```

```python
from backports import ssl
import imapclient

context = ssl.SSLContext(ssl.PROTOCOL_TLSv1_2)

imapObj = imapclient.IMAPClient('imap.gmail.com', ssl=True, ssl_context=context)
```

```python
    # 16.4.2 登录到IMAP服务器
```

```python
imapObj.login('enzolife.sife.sysu@gmail.com','nzqbnevdwzwthopp')
```

```python
    # 16.4.3 搜索电子邮件
```

```python
    # 16.4.4 选择文件夹
```

```python
import pprint
pprint.pprint(imapObj.list_folders())
```

```python
imapObj.select_folder('INBOX', readonly=True)
```

```python
    # 16.4.5 执行搜索
```

```python
Amazon_email = imapObj.search([u'From', u'Amazon',u'SINCE',datetime.datetime(2017, 6, 13)])
Amazon_email
```

```python
rawMessages = imapObj.fetch(Amazon_email,['BODY[]'])
import pprint
pprint.pprint(rawMessages)
```

```python
    # 16.4.8 从原始信息中获取电子邮件地址
```

```python
import pyzmail
message = pyzmail.PyzMessage.factory(rawMessages[18860][b'BODY[]'])
```

```python
message.get_subject()
```

```python
message.get_addresses('from')
```

```python
message.get_addresses('to')
```

```python
message.get_addresses('cc')
```

```python
message.get_addresses('bcc')
```

```python
    # 16.4.9 从原始消息中获取正文
```

```python
message.html_part != None
```

```python
message.html_part.get_payload().decode(message.html_part.charset)
```
