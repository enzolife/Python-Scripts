```python
# 开始录入信息
# 读取商品信息表
product_detail_path = 'D://Program Files (x86)//百度云同步盘//Dropbox//-E·J- 2014.5.1//2015.2.18 速卖通//2016.11.26 速卖通数据分析//2016.12.18 速卖通商品分析.xlsm'
product_detail_df = pd.read_excel(product_detail_path,  sheetname = '商品明细表')
product_detail_df
```

```python
# 读取商品属性表
product_parameter_df = pd.read_excel(product_detail_path, sheetname = '商品属性')
product_parameter_df
```

```python
# 两表合并
product_detail_parameter_df = pd.merge(product_detail_df, product_parameter_df, how='left', left_on='商品ID', right_on='商品ID')
product_detail_parameter_df
```

```python
# 取EAN表
product_ean_path = "D://Program Files (x86)//百度云同步盘//Dropbox//-E·J- 2014.5.1//2017.4.7 Cdiscount//2017.11.29 新产品创建//2017.12.1 Cdiscount EAN对照表.xlsx"
product_ean_df = pd.read_excel(product_ean_path)
product_ean_df
```

```python
# 合并到大表
product_detail_parameter_ean_df = pd.merge(product_detail_parameter_df, product_ean_df, how='inner', left_on='商品ID', right_on='商品ID')
product_detail_parameter_ean_df
```

```python
product_detail_parameter_ean_df[:1].unstack()
```

```python
# 开始输入商品信息
for index, product in product_detail_final[36:].iterrows():
    # 点击Stock
    click_stock = browser.find_elements_by_css_selector('.dropdown-toggle')
    time.sleep(5)
    click_stock[2].click()
    # 点击create a product
    create_a_product = browser.find_element_by_link_text('Create a product')
    create_a_product.click()
    time.sleep(30)
    # 选择1级分类
    cat_1_select = browser.find_elements_by_css_selector('.jsCategoryName')
    cat_1_select[11].click()
    time.sleep(5)
    # 选择2级分类
    cat_2_select = browser.find_elements_by_css_selector('.jsCategoryName')
    cat_2_select[69].click()
    time.sleep(5)
    # 选择3级分类
    cat_3_select = browser.find_elements_by_css_selector('.jsCategoryName')
    cat_3_select[93].click()
    time.sleep(5)
    # 点击Okay
    submit_cat_button = browser.find_element_by_css_selector('.btn.small.btn-primary')
    submit_cat_button.click()
    time.sleep(5)
    # 录入信息
    input_box_elem = browser.find_elements_by_css_selector('.form-control')
    '''
    for i in range(len(input_box_elem)):
        print(input_box_elem[i].get_attribute('id'))
    '''
    input_data_list = [['商品ID', 0],
                         ['Product Name in French', 1],
                         ['EAN Number Final', 2],
                         ['Product Description in French', 4],
                         ['附图1（URL）地址', 5],
                         ['Product Description in French', 11],
                         ['附图2（URL）地址', 12],
                         ['附图3（URL）地址', 14],
                         ['附图4（URL）地址', 16]]
    
    for input_data in input_data_list:
        input_data_name = input_data[0]
        input_data_location = input_data[1]
        
        input_box_elem[input_data_location].click()
        time.sleep(5)
        input_box_elem[input_data_location].send_keys(product[input_data_name])
        
    
    # 点击创建产品
    create_a_product = browser.find_elements_by_css_selector('.btn.btn-primary')
    create_a_product[3].click()
    
    # 提示语
    print(str(product['商品ID']) + ' ' + product['Product Name in French'] + ' is created.')
    time.sleep(30)
```
