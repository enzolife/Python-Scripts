```python
from Get_Google_Sheets import get_certain_google_sheets_to_dataframe
from Get_Particular_Date import *
import pandas as pd
import numpy as np
```

```python
book_name = 'Seller_Index_from_Salesforce'
select_sheet = 'Raw_Seller_Index'
seller_index = get_certain_google_sheets_to_dataframe(book_name, select_sheet)
```

```python
# 修改日期格式
seller_index['GP Account Shopee Account Created Date'] \
    = pd.to_datetime(seller_index['GP Account Shopee Account Created Date'], format='%d/%m/%Y')
# 昨天
seller_index['Yesterday Date'] = get_yesterday_date()
seller_index
```

```python
seller_index.info()
```

```python
gp_pre_index = seller_index[['GP Account ID',
                                 'GP Account Name',
                                 'GP Account Owner',
                                 'GP Account Shopee Account Created Date']]
gp_pre_index = gp_pre_index.drop_duplicates()
gp_pre_index
```

```python
gp_pre_index['Yesterday Date'] = get_yesterday_date()
gp_pre_index['Yesterday Date'] = pd.to_datetime(gp_pre_index['Yesterday Date'])
gp_pre_index['GP Acc Duration'] = gp_pre_index['Yesterday Date'] - gp_pre_index['GP Account Shopee Account Created Date']
gp_pre_index
```

```python
gp_pre_index['GP Acc Duration'] = gp_pre_index['GP Acc Duration'] / np.timedelta64(1, 'D')
gp_pre_index
```

```python
gp_pre_index['GP Acc Creaated at M-1'] = np.where((gp_pre_index['GP Account Shopee Account Created Date']<='2017-07-31')&(gp_pre_index['GP Account Shopee Account Created Date']>='2017-07-01'), 1, 0)
gp_pre_index['GP Acc Creaated at M-2'] = np.where((gp_pre_index['GP Account Shopee Account Created Date']<='2017-06-30')&(gp_pre_index['GP Account Shopee Account Created Date']>='2017-06-01'), 1, 0)
gp_pre_index['GP Acc Creaated at M-3'] = np.where((gp_pre_index['GP Account Shopee Account Created Date']<='2017-05-31')&(gp_pre_index['GP Account Shopee Account Created Date']>='2017-05-01'), 1, 0)
gp_pre_index
```

```python
gp_pre_index_group = gp_pre_index.groupby([seller_index['GP Account Owner']])
gp_pre_index_result = gp_pre_index_group.agg({'GP Acc Creaated at M-1': 'sum',
                                                  'GP Acc Creaated at M-2': 'sum',
                                                  'GP Acc Creaated at M-3': 'sum'})
gp_pre_index_result = pd.DataFrame(gp_pre_index_result).reset_index()
gp_pre_index_result
```

```python
def get_start_end_of_certain_month(months_to_minus, start_or_end):

    compared_date = get_yesterday_date()
    i = 0
    while i < months_to_minus:
        end_of_certain_month = datetime.date(compared_date.year, compared_date.month, 1) + datetime.timedelta(days=-1)
        start_of_certain_month = datetime.date(end_of_certain_month.year, end_of_certain_month.month, 1)
        compared_date = start_of_certain_month
        i = i + 1

    if start_or_end == 'start':
        return start_of_certain_month
    elif start_or_end == 'end':
        return end_of_certain_month

print(get_start_end_of_certain_month(1, "end"))
print(get_start_end_of_certain_month(1, "start"))
```
