```python
# for product_id in product_list['商品ID'][:40]:
for product_id in product_list['商品ID'][40:56]:
    # 切换回主界面
    browser.switch_to.default_content()
    # 切换到输入product id的iframe
    browser.switch_to.frame(browser.find_elements_by_xpath("//iframe[contains(@src,'productSelect')]")[-1])
    # 输入productid
    product_id_input_box = browser.find_elements_by_css_selector(".input-product.txt-tip")[0]
    product_id_input_box.clear()
    product_id_input_box.send_keys(int(product_id))
    # 点击搜索
    try:
        product_search_button = browser.find_element_by_css_selector(".ui-button.ui-button-primary.ui-button-small")
        product_search_button.click()
        time.sleep(5)
        # 点击全选
        select_all_box = browser.find_element_by_id("checkbox-all")
        select_all_box.click()
        time.sleep(5)
    except Exception as err:
        print("Can't find this product: " + str(err))
        pass

# submit_button_click = browser.find_elements_by_css_selector(".ui-button ui-button-primary.ui-button-medium")[-1]
# submit_button_click.click()
```
