```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
import zipfile
import shutil
```

```python
# 屏幕最大化，且指定下载目录
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

prefs = {"profile.default_content_settings.popups": 0,
         "download.default_directory": r"D:\Program Files (x86)\百度云同步盘\Dropbox\-E·J- 2014.5.1\2016.12.15 店小秘数据分析\\", # IMPORTANT - ENDING SLASH V IMPORTANT
         "directory_upgrade": True}
options.add_experimental_option("prefs", prefs)
```

```python
# 使用chromedriver才可以用开发者权限
chrome_driver_path = ".//chrome_driver//chromedriver.exe"
browser = webdriver.Chrome(chrome_driver_path, chrome_options=options)
```

```python
browser.get('https://merchant.factorymarket.com/products/card/8dece647-a0b0-4193-8427-9beb05d2ec25')
time.sleep(5)
```

```python
# active / inactive 
product_active = browser.find_elements_by_css_selector('.ant-checkbox.ant-checkbox-checked')

if len(product_active) == 0:
    product_inactive = browser.find_elements_by_css_selector('.ant-checkbox')[0]
    product_inactive.click()
    time.sleep(5)
    
body = browser.find_element_by_css_selector('body')
body.send_keys(Keys.END)

product_updates = browser.find_elements_by_css_selector('.ant-btn.ant-btn-primary')[-1]
product_updates.click()
time.sleep(5)
```

```python
for index, product in factory_mart_product_list.iterrows():
    product_url = product['url']
    # logging.info('product_url: {}'.format(product_url))
    
    browser.get(product_url)
    time.sleep(5)

    # active / inactive 
    product_active = browser.find_elements_by_css_selector('.ant-checkbox.ant-checkbox-checked')

    if len(product_active) == 0:
        product_inactive = browser.find_elements_by_css_selector('.ant-checkbox')[0]
        product_inactive.click()
        time.sleep(5)

    body = browser.find_element_by_css_selector('body')
    body.send_keys(Keys.END)

    product_updates = browser.find_elements_by_css_selector('.ant-btn.ant-btn-primary')[-1]
    product_updates.click()
    time.sleep(5)
    logging.info('{} has been updated.'.format(factory_mart_product_list['product_name']))
```

```python
for index, product in factory_mart_product_list.iterrows():
    product_url = product['url']
    # logging.info('product_url: {}'.format(product_url))
    product_name_germany = product['product_name_germany']
    
    browser.get(product_url)
    time.sleep(5)

    # product_name
    product_name = browser.find_elements_by_css_selector('.ant-input')[0]
    product_name.clear()
    time.sleep(5)
    product_name.send_keys(product_name_germany)
    
    body = browser.find_element_by_css_selector('body')
    body.send_keys(Keys.END)

    product_updates = browser.find_elements_by_css_selector('.ant-btn.ant-btn-primary')[-1]
    product_updates.click()
    time.sleep(5)
    logging.info('{} has been updated.'.format(product['product_name']))
```
