```python
from selenium import webdriver
import pandas as pd
import time
```

```python
# 屏幕最大化，且指定下载目录
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

prefs = {"profile.default_content_settings.popups": 0,
         "directory_upgrade": True}
options.add_experimental_option("prefs", prefs)
```

```python
# 使用chromedriver才可以用开发者权限
chrome_driver_path = "D://Program Files (x86)//百度云同步盘//我的软件//chromedriver.exe"
browser = webdriver.Chrome(chrome_driver_path, chrome_options=options)
```

```python
type(browser)
```

```python
# 先读取Top Product列表
top_product_path = "D://Program Files (x86)//FangCloudV2//寶寶優選//寶寶優選公共资料库//02 【虾皮·印尼站】tengus.id//2018.4.21 印尼站母婴跨境产品销量列表.xlsx"
top_product = pd.read_excel(top_product_path)
top_product.head()
```

```python
# 创建空的list
product_main_img_url_list = []

for i, product in top_product.iterrows():
    try:
        browser.get(product["product_url"])
    
        product_main_img = browser.find_element_by_css_selector('.shopee-image-gallery__main-image')
        product_main_img_url = product_main_img.get_attribute('style')
        
        product_main_img_url = product_main_img_url.split('\"')[1]
        print(product_main_img_url)
        
        product_main_img_url_list.append(product_main_img_url)
        
        print(str(i + 1) + ' product is captured.')
    except Exception as err:
        print("An exception occurred: " + str(err))
        product_main_img_url_list.append("not captured")

# 转换为data frame
# product_main_img_url_data_frame = pd.DataFrame({'product_main_img_url': product_main_img_url_list})
# 加入到top_product
top_product['product_main_img_url'] = product_main_img_url_list
        
print(product_main_img_url_list)
```
