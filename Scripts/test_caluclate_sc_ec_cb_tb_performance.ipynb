```python
import pandas as pd
import os
```

```python
file_folder = "D:\\Program Files (x86)\\百度云同步盘\\我的文档"
```

```python
file_path = os.path.join(file_folder, 'EC_TB.xlsx')
```

```python
file_content = pd.read_excel(file_path)
file_content.head()
```

```python
file_content = file_content[file_content['purchased_on'] >= '2018-10-15']
```

```python
result_group = file_content.groupby(['purchased_on'])
```
