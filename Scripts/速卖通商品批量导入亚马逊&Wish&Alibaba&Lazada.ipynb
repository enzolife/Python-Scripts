```python
# 导入速卖通折扣信息
import os
smt_discount_list = []

# 查找文件夹里面的ZIP file
file_folder = "C:\\Users\\kuang\\Downloads"
for filename in os.listdir(file_folder):
    if filename.startswith('速卖通折扣') and filename.endswith('.xlsx'):
        filepath = os.path.join(file_folder, filename)
        smt_discount = pd.read_excel(filepath)
        smt_discount_list.append(smt_discount)

smt_discount_df = pd.concat(smt_discount_list)
smt_discount_df.head()
```

```python
df_new_instance = df
```

```python
# join 速卖通折扣
df_new_instance = pd.merge(df_new_instance, smt_discount_df, how='left', left_on='平台产品ID', right_on='Product ID')
df_new_instance.head()
```

```python
# 无折扣产品写成0
df_new_instance['Discount'] = df_new_instance['Discount'].fillna(0)
df_new_instance.head()
```

```python
# 生成亚马逊需要的字段
df_new_instance['item_sku'] = df_new_instance['平台产品ID']
df_new_instance['brand_name'] = 'Woli'
df_new_instance['external_product_id'] = ''
df_new_instance['external_product_id_type'] = 'UPC'
df_new_instance['item_name'] = df_new_instance['产品名称'].str.capitalize()
df_new_instance['manufacturer'] = 'Guangzhou Woli Trading Co. Ltd.'
df_new_instance['quantity'] = 10000
```

```python
# 拆图片链接
df_new_instance['main_image_url'] = df_new_instance['产品图片'].str.split(';').str[0]
df_new_instance['other_image_url1'] = df_new_instance['产品图片'].str.split(';').str[1]
df_new_instance['other_image_url2'] = df_new_instance['产品图片'].str.split(';').str[2]
df_new_instance['other_image_url3'] = df_new_instance['产品图片'].str.split(';').str[3]
df_new_instance['other_image_url4'] = df_new_instance['产品图片'].str.split(';').str[4]
df_new_instance['other_image_url5'] = df_new_instance['产品图片'].str.split(';').str[5]
df_new_instance['other_image_url6'] = df_new_instance['产品图片'].str.split(';').str[6]
df_new_instance['other_image_url7'] = df_new_instance['产品图片'].str.split(';').str[7]
df_new_instance['other_image_url8'] = df_new_instance['产品图片'].str.split(';').str[8]

df_new_instance
```

```python
# 价格处理
# df_new_instance['price_original'] = ''

import json
from pandas.io.json import json_normalize

# df_new_instance['价格信息'].apply(json.loads)
```

```python
# 其他属性
df_new_instance['update_delete'] = 'Update'
df_new_instance['product_description'] = ''

df_new_instance['Manufacturer Part Number'] = df_new_instance['平台产品ID']

df_new_instance['bullet_point1'] = 'Output Voltage: '
df_new_instance['bullet_point2'] = 'Output Wattage: '
df_new_instance['bullet_point3'] = 'Output Current: '
df_new_instance['bullet_point4'] = 'Brand: Woli'
df_new_instance['bullet_point5'] = 'Estimated Delivery Time: 5-10 days (expedited); 10-15 days (Standard)'

df_new_instance['generic_keywords'] = 'laptop charger'

df_new_instance['color_name'] = 'black'
df_new_instance['color_map'] = 'Black'

df_new_instance['voltage'] = 0
df_new_instance['voltage_unit_of_measure'] = 'volts'
df_new_instance['wattage'] = 0
df_new_instance['wattage_unit_of_measure'] = 'Watts'

df_new_instance['item_length'] = 18
df_new_instance['item_width'] = 13
df_new_instance['item_height'] = 5

df_new_instance['item_length_unit_of_measure'] = 'CM'
df_new_instance['item_height_unit_of_measure'] = 'CM'
df_new_instance['item_width_unit_of_measure'] = 'CM'

df_new_instance['country_of_origin'] = 'China'

df_new_instance
```

```python
# 重量&运费&价格
```

```python
usd_to_rmb_currency = 6.50
usd_to_gbp_currency = 0.73
usd_to_eur_currency = 0.85
usd_to_sek_currency = 8.81
usd_to_pln_currency = 3.92
usd_to_aud_currency = 1.40
usd_to_try_currency = 9.03

df_new_instance['运费'] = np.where(df_new_instance['产品包装后的重量']<0.3, 30, df_new_instance['产品包装后的重量']*100)
df_new_instance['运费USD'] = df_new_instance['运费'] / usd_to_rmb_currency
```

```python
# 部分产品速卖通产品没有含运费，需要按需计算
df_new_instance['产品实价'] = np.select(
    [
        (df_new_instance['运费模板'] == 717975420),
        (df_new_instance['运费模板'] == 702140041)
    ],
    [
        (df_new_instance['产品平台定价']),
        (df_new_instance['产品平台定价'] - df_new_instance['运费USD'])
    ],
    default=0
)
df_new_instance.head()
```

```python
df_new_instance['产品平台定价+运费USD'] = df_new_instance['产品实价'] + df_new_instance['运费USD']

# 佣金15%，其他流量费等10%，利润10%
df_new_instance['含佣总价'] = df_new_instance['产品平台定价+运费USD'] / (1-0.35)

# 其他国家定价
df_new_instance['含佣总价GBP'] = round(df_new_instance['含佣总价'] * usd_to_gbp_currency,1)
df_new_instance['含佣总价EUR'] = round(df_new_instance['含佣总价'] * usd_to_eur_currency,1)
df_new_instance['含佣总价SEK'] = round(df_new_instance['含佣总价'] * usd_to_sek_currency,1)
df_new_instance['含佣总价PLN'] = round(df_new_instance['含佣总价'] * usd_to_pln_currency,1)
df_new_instance['含佣总价AUD'] = round(df_new_instance['含佣总价'] * usd_to_aud_currency,1)
df_new_instance['含佣总价TRY'] = round(df_new_instance['含佣总价'] * usd_to_try_currency,1)

df_new_instance
```
