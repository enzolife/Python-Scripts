```python
import pandas as pd
import os
import numpy as np
```

```python
folder_path = 'D:\\Program Files (x86)\\百度云同步盘\\Dropbox\\Shopee 2016.4.12\\2016.8.25 Temp Report Folder'
file_name = 'Python_get_payout_channel_distribution_2018_07_25.xlsx'
```

```python
file_path = os.path.join(folder_path, file_name)
```

```python
payout_channel_distribution_df = pd.read_excel(file_path)
```

```python
payoneer_list = payout_channel_distribution_df[payout_channel_distribution_df['name'] == 'Payoneer Payout']
payoneer_list
```

```python
payoneer_list_group = payoneeer_list.groupby(['payoneer_account_email'])
payoneer_list_result = payoneer_list_group.agg({'gp_account_name': 'nunique'}).reset_index().rename(columns={"gp_account_name": "# of unique GP account"})

payoneer_list_result =  payoneer_list_result[payoneer_list_result['# of unique GP account'] > 1]

payoneer_list_result.sort_values(by=['# of unique GP account'], ascending=[False])
```

```python
payoneer_list_with_one_more_gp = payoneer_list[payoneer_list['payoneer_account_email'].isin(payoneer_list_result['payoneer_account_email'])]
payoneer_list_with_one_more_gp
```

```python
payoneer_list_with_one_more_gp_group = payoneer_list_with_one_more_gp.groupby(['gp_account_name', 'payoneer_account_email'])
payoneer_list_with_one_more_gp_result = payoneer_list_with_one_more_gp_group.agg({'mtd_ado': 'sum'}).reset_index()
payoneer_list_with_one_more_gp_result.sort_values(by=['gp_account_name', 'mtd_ado'], ascending=[False, False])
```

```python
payoneer_list_with_one_more_gp_result_email = payoneer_list_with_one_more_gp_result['payoneer_account_email'].drop_duplicates()
payoneer_list_with_one_more_gp_result_email = payoneer_list_with_one_more_gp_result_email.reset_index()
payoneer_list_with_one_more_gp_result_email
```

```python
payoneer_list_with_one_more_gp_result = pd.merge(payoneer_list_with_one_more_gp_result, payoneer_list_with_one_more_gp_result_email, on=['payoneer_account_email'])
payoneer_list_with_one_more_gp_result
```

```python
payoneer_list_with_one_more_gp_result = payoneer_list_with_one_more_gp_result.sort_values(by=['index', 'mtd_ado'], ascending=[True, False])
payoneer_list_with_one_more_gp_result['group_sort'] = payoneer_list_with_one_more_gp_result['mtd_ado'].groupby(payoneer_list_with_one_more_gp_result['index']).rank(ascending=0, method='dense')
payoneer_list_with_one_more_gp_result
```

```python
payoneer_email_primay_list = payoneer_list_with_one_more_gp_result[payoneer_list_with_one_more_gp_result['group_sort'] == 1]
payoneer_email_primay_list
```
