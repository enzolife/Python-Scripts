```python
# 读取商品价格表
product_price_path = "D://Program Files (x86)//百度云同步盘//Dropbox//-E·J- 2014.5.1//2017.4.7 Cdiscount//2017.11.29 新产品创建//2017.12.5 成功上传的商品表.xlsx"
product_price_list = pd.read_excel(product_price_path,  sheetname = 'Final')
product_price_list
```

```python
for index, product in product_price_list[1:].iterrows():
    # 点击Stock
    click_stock = browser.find_elements_by_css_selector('.dropdown-toggle')
    click_stock[2].text
    time.sleep(5)
    click_stock[2].click()    
    # 点击add an offer
    add_an_offer = browser.find_element_by_link_text('Add an offer')
    # create_a_product.get_attribute('href')
    add_an_offer.click()
    # 输入ean id
    search_product_input_box = browser.find_element_by_css_selector('.form-control')
    search_product_input_box.click()
    search_product_input_box.send_keys(product['Gencod'])
    time.sleep(5)
    # 点击搜索
    search_product_button = browser.find_elements_by_css_selector('.btn.btn-primary')
    search_product_button[3].click()
    time.sleep(5)
    # 点击sell yours
    sell_yours_button = browser.find_element_by_css_selector('.btn.sndLink.btn-default.hidden-xs')
    sell_yours_button.click()
    time.sleep(5)
    # 输入各项信息，依次为商品ID，售价，库存，发货期
    input_box = browser.find_elements_by_css_selector('.formInputText.borRad5.liveValidatedField')
    input_box[0].send_keys(product['商品ID'])
    input_box[2].clear()
    input_box[2].send_keys(product['折后价欧元'])
    input_box[5].clear()
    input_box[5].send_keys(1000)
    input_box[18].send_keys(3)
    
    # 输入物流费信息
    shipping_input_box = browser.find_elements_by_css_selector('.inputSearchQte')
    shipping_input_box[0].send_keys(0)
    shipping_input_box[1].send_keys(0)
    shipping_input_box[2].send_keys(0)
    shipping_input_box[3].send_keys(0)
    
    # 提交
    submit_button = browser.find_element_by_css_selector('.positiveButtonExpand.floatRight.borRad5')
    time.sleep(10)
    submit_button.click()
    
    print(str(product['商品ID']) + ' is added an offer.')
```
